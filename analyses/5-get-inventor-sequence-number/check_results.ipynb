```python
import pandas as pd
import numpy as np
```

```python
#read dataset
disam = pd.read_csv('data/eval_als.txt', sep='\t', header=None, names = ['code', 'inventor'], dtype={'code': 'string'})

#separate code into patent and sequence
disam['patent'] = disam.apply(lambda row: int(row.code[0:7]), axis=1)
disam['sequence'] = disam.apply(lambda row: int(row.code[8: len(row.code)]), axis=1)

disam = disam.drop(['code'], axis=1)

disam.head(10)
```

```python
#read in dataset
results = pd.read_csv('data/patents_2005_012_autosequence.csv', index_col=0)
results['sequence'] = pd.to_numeric(results['sequence'], errors='coerce').astype(pd.Int16Dtype())
results.set_index(['patent', 'sequence'], inplace=True)
results.head(10)
```

```python
join = disam.join(results, on=['patent', 'sequence'], how='inner')
join.head(10)
```

```python
join.to_csv('join.csv')
```
